```python
import pandas as pd

# Read data from CSV
df = pd.read_csv("D:\input (3).csv")

# Sort data by employee ID and date
df['Date of Joining'] = pd.to_datetime(df['Date of Joining'])
df.sort_values(by=['Employee Code', 'Date of Joining'], inplace=True)

# Calculate 'Effective Date' and 'End Date'
df['Effective Date'] = df.groupby('Employee Code')['Date of Joining'].shift(-1, fill_value=pd.Timestamp.today())
df['End Date'] = df['Effective Date'].shift(-1, fill_value=pd.Timestamp('2100-01-01')) - pd.Timedelta(days=1)

# Calculate Last Compensation
df['Last Compensation'] = df.groupby('Employee Code')['Compensation'].shift(1)

# Choose the latest compensation value between 'Compensation', 'Compensation 1', and 'Compensation 2'
df['Compensation'] = df[['Compensation', 'Compensation 1', 'Compensation 2']].iloc[:, -1]

# Calculate Tenure in Org
df['Tenure in Org'] = (df['Effective Date'] - df['Date of Joining']).dt.days / 365

# Create Variable Pay column and fill missing values
df['Variable Pay'] = None  # Create a new column filled with None (NaN)

# Create 'Last Pay Raise Date' by shifting 'Compensation 1 date' column
df['Last Pay Raise Date'] = df.groupby('Employee Code')['Compensation 1 date'].shift(1)

# Create 'Performance Rating' column using 'Review 1' and 'Review 2'
df['Performance Rating'] = df[['Review 1', 'Review 2']].max(axis=1)

# Choose the latest engagement score between 'Engagement 1' and 'Engagement 2'
df['Engagement Score'] = df[['Engagement 1', 'Engagement 2']].max(axis=1)

# Select desired columns for output
output_df = df[['Employee Code', 'Manager Employee Code', 'Last Compensation', 'Compensation',
                'Last Pay Raise Date', 'Variable Pay', 'Tenure in Org', 'Performance Rating',
                'Engagement Score', 'Effective Date', 'End Date']]

# Save transformed data to CSV
output_df.to_csv("D:\output.csv", index=False)
```
